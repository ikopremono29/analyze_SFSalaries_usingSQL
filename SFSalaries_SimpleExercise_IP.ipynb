First thing first, we import pandas and sqlite3 package to be able to run SQL query within this notebook

```python
import pandas as pd
import sqlite3 as sql
```

Then, we connect to the database. We have a database named 'SF_salaries_database.sqlite'.

```python
db = "SF_salaries_database.sqlite"
conn = sql.connect(db)
```

Step 1: Getting to know the table

First thing first, quite important to have the feeling of the table. Let's see how many rows and column this table has.

```python
query0 = '''SELECT * FROM Salaries
            ORDER BY TotalPay DESC
'''
```

```python
df0 = pd.read_sql_query(query0,conn)
df0
```

We got 148654 rows and 13 columns from the table.
Look at some 'strange' data in last 5 rows. We focus at rows with no value there.
Let's filter it in Step 2.

Step 2: Filtering unwanted data

Here we'd like to exclude some unwanted data which contains 'Not provided' in there.

```python
query1 = '''SELECT * FROM Salaries
            WHERE EmployeeName IS NOT 'Not provided' OR JobTitle IS NOT 'Not provided'
            ORDER BY TotalPay DESC
'''
```

```python
df1 = pd.read_sql_query(query1, conn)
df1
```

4 rows are filtered out.
We proceed step 3 to see who is the highest earner.
The highest earner we get from TotalPayBenefits column.

We try to limit to 10 rows only.

Step 3: See top 10 of the highest earner and sort it out from the highest to the least.

```python
query2 = '''SELECT * FROM Salaries
            ORDER BY TotalPayBenefits DESC
            LIMIT 10
'''
```

```python
df2 = pd.read_sql_query(query2, conn)
df2
```

We see that Nathaniel Ford and Gary Jimenez are among the top earner in San Fransisco.
But their records are in the same year.

Let's who are the top earner each year in Step 4.

Step 4: Seek out top earner employee each year

```python
query3 = '''SELECT Year, EmployeeName, JobTitle, MAX(TotalPayBenefits) FROM Salaries
            WHERE EmployeeName IS NOT 'Not provided' OR JobTitle IS NOT 'Not provided' OR TotalPayBenefits > 0
            GROUP BY Year
            LIMIT 10
'''
```

```python
df3 = pd.read_sql_query(query3, conn)
df3
```

Now, we can see who are the top earner for each year, from 2011 to 2014
However, we don't see any engineer job title in those list. I'm curious how much engineers in this dataset.

Step 5. Find Total Earning for Engineer Job Title

```python
query4 = '''SELECT EmployeeName, JobTitle, TotalPayBenefits, Year FROM Salaries
            WHERE JobTitle LIKE 'Engine%' AND EmployeeName IS NOT 'Not provided'
            ORDER BY TotalPayBenefits DESC
            LIMIT 10
'''
```

```python
df4 = pd.read_sql_query(query4, conn)
df4
```

We find that top 10 engineers are earning between 240000 to 250000 anually.
Finally, in this last step, let's create a query that seek out what are the job titles in San Fransisco with top average total earning.

Step 6:
In this step, we use window function in order to see the rank for the job title sorted from the highest max total pay.

```python
query5 = '''SELECT RANK() OVER(ORDER BY MAX(TotalPayBenefits) DESC) AS 'Rank by Max Total Pay',            
            JobTitle, ROUND(AVG(TotalPayBenefits),1) AS 'Average Total Pay', Year FROM Salaries
            WHERE EmployeeName IS NOT 'Not provided' OR JobTitle IS NOT 'Not provided'
            GROUP BY JobTitle
            HAVING AVG(TotalPayBenefits) > 350000
            ORDER BY 3 DESC
'''
```

```python
df5 = pd.read_sql(query5, conn)
df5
```

We got General Manager-Metropolitan Transit Authority ranks 1 in total pay but lies in no 4 in average total pay. The top average total pay is Chief Investment Office with 436224.4 annual earning.
